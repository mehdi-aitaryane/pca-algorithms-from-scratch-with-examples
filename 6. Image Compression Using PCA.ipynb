# Image Compression Using PCA

```python
import matplotlib.pyplot as plt
import numpy as np
from models import PCA
from sklearn.datasets import fetch_olivetti_faces   # face data set (400 faces image)
```

```python
faces = fetch_olivetti_faces()
```

```python
img = faces.images[50]
# Create a figure with subplots
fig, axes = plt.subplots(2, 3, figsize=(15, 8))

# Display the original image
axes[0, 0].imshow(img, cmap="gray")
axes[0, 0].set_title("Original Image (600 Components)")

# Display compressed images with varying percentages
percentages = [64, 50, 35, 20, 5]
indexs = [[0, 1], [0, 2], [1, 0], [1, 1], [1, 2]]
for i, percentage in enumerate(percentages):
    tswizzle_pca = PCA(n_components=percentage)
    transformed = tswizzle_pca.fit_transform(img)
    projected = tswizzle_pca.inverse_transform(transformed)
    axes[indexs[i][0], indexs[i][1]].imshow(projected, cmap="gray")
    axes[indexs[i][0], indexs[i][1]].set_title(f"Compressed Image ({percentage} Components)")

# Adjust layout for better viewing
plt.tight_layout()

# Display the plot
plt.show()
```
