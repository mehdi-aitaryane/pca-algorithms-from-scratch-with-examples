# Image Denoising Using Kernel PCA

```python
import matplotlib.pyplot as plt
import numpy as np
from models import KernelPCA
from sklearn.datasets import fetch_olivetti_faces   # face data set (400 faces image)
```

```python
faces = fetch_olivetti_faces()
```

```python
facesNoise = faces.data + np.random.normal(loc=0.0, scale=0.05, size=faces.data.shape)

pca = KernelPCA(n_components=0.9999)
pca.fit(facesNoise)
pcaFaces = pca.transform(facesNoise)
facesRecons = pca.inverse_transform(pcaFaces)
```

```python
fig, ax = plt.subplots(figsize=(10,6), nrows=1, ncols=3)

ax[0].set_title('Original Image')
ax[0].imshow(np.reshape(faces.data[18], (64,64)), cmap='gray')
ax[0].axis('off')

ax[1].set_title('Noise Image')
ax[1].imshow(np.reshape(facesNoise[18], (64,64)), cmap='gray')
ax[1].axis('off')

ax[2].set_title('Kernel PCA Image')
ax[2].imshow(np.reshape(facesRecons[18], (64,64)), cmap='gray')
ax[2].axis('off')

plt.show()
```
