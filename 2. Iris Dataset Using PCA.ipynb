# Iris Dataset Using PCA

## Imports

```python
from datasets import count_occurances, print_shape
from sklearn.datasets import load_iris
from plots import scatter2D, scatter3D
from models import PCA
```

## Initialize X and y

```python
X, y = load_iris().data, load_iris().target
label_names = list(load_iris().target_names)
```

## Print Shape

```python
print_shape(X, y)
```

## Count Occurances

```python
count_occurances(y, label_names=label_names)
```

## Initilize PCA with n_components = 2

```python
pca = PCA(n_components=2)
X_T = pca.fit_transform(X)
```

## Plot 2D

```python
scatter2D(title="2D Iris Dataset", xlabel="PC1", ylabel="PC2", X = X_T, y = y, label_names=label_names)
```

## Initilize PCA with n_components = 3

```python
pca = PCA(n_components=3)
X_T = pca.fit_transform(X)
```

## Scatter 3D

```python
scatter3D(title="3D Iris Dataset", xlabel="PC1", ylabel="PC2",  zlabel="PC3", X = X_T, y = y, label_names = label_names)
```
