# Circles Dataset Using PCA

## Imports

```python
from datasets import count_occurances, print_shape
from sklearn.datasets import make_circles
from plots import scatter2D
from models import PCA
```

## Initialize X and y

```python
X, y = make_circles(n_samples=1000, factor=0.3, noise=0.02, random_state=42)
```

## Print Shape

```python
print_shape(X, y)
```

## Count Occurances

```python
count_occurances(y)
```

## Scatter 2D

```python
scatter2D(title="2D Circles Dataset", xlabel="X", ylabel="Y", X = X, y = y)
```

## Initilize PCA with n_components = 2

```python
pca = PCA(n_components=2)
X_T = pca.fit_transform(X)
```

## Scatter 2D

```python
scatter2D(title="2D Circles Dataset", xlabel="PC1", ylabel="PC2", X = X_T, y = y)
```
