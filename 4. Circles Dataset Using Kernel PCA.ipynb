# Circles Dataset Using Kernel PCA

## Imports

```python
from datasets import count_occurances, print_shape
from sklearn.datasets import make_circles
from plots import scatter2D, scatter3D
from models import KernelPCA
from kernels import RBFKernel
```

## Initialize X and y

```python
X, y = make_circles(n_samples=1000, factor=0.3, noise=0.02, random_state=42)
```

## Print Shape

```python
print_shape(X, y)
```

## Count Occurances

```python
count_occurances(y)
```

## Scatter 2D

```python
scatter2D(title="2D Circles Dataset", xlabel="X", ylabel="Y", X = X, y = y)
```

## Initilize PCA with n_components = 2

```python
kernel = RBFKernel(gamma=1.0)
pca = KernelPCA(n_components=2, alpha=1.0, kernel=kernel)
X_T = pca.fit_transform(X)
```

## Plot 2D

```python
scatter2D(title="2D Circles Dataset", xlabel="PC1", ylabel="PC2", X = X_T, y = y)
```

## Initilize PCA with n_components = 3

```python
kernel = RBFKernel(gamma=1.0)
pca = KernelPCA(n_components=3, alpha=1.0, kernel=kernel)
X_T = pca.fit_transform(X)
```

## Scatter 3D

```python
scatter3D(title="3D Circles Dataset", xlabel="PC1", ylabel="PC2",  zlabel="PC3", X = X_T, y = y)
```
